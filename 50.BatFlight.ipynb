```python
%matplotlib inline

""" Example program with animation with a bat """

import numpy as np
from matplotlib import pyplot as plt
from matplotlib import animation, rc

rc('text', usetex=True)
params = {'text.latex.preamble': [r'\usepackage{marvosym}']}
plt.rcParams.update(params)

```

```python
from math import atan2, degrees
def bat_flight(frames=360, interval=30, dt=0.3):
    fig = plt.figure(figsize=(8, 6))
    ax = plt.axes(xlim=(0, 10), ylim=(0, 10))

    # Animated.
    bat = ax.text(8, 5, r"\Bat", size=32, ha='center', va='center')

    def init():
        bat.set_position((8, 5))
        return bat

    def animate(i):
        x = 5 + 3 * np.sin(np.radians(i))
        y = 5 + 3 * np.cos(np.radians(i))

        bat.set_position((x, y))
        alpha = degrees(atan2(-y,x))
        bat.set_rotation(alpha)
        
        return bat

    return animation.FuncAnimation(fig, animate, init_func=init,
                                   frames=frames, interval=interval)
```
